```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
df
```

```python
plt.figure(figsize=(6,7))
plt.scatter(df['year'],df['per capita income (US$)'])
plt.xlabel("Year")
plt.ylabel("Per capita income")
plt.title("Per capita income over the Years")
plt.show()
```

```python
new_df=df.drop("per capita income (US$)",axis="columns")
new_df
```

```python
income=df['per capita income (US$)']
income
```

```python
reg=linear_model.LinearRegression()
reg.fit(new_df,income)
```

```python
reg.predict([[2020]])
```

```python
reg.predict([[2019]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.plot(df['year'], df['per capita income (US$)'],label='Actual per Capita')
plt.xlabel("Year")
plt.ylabel("Per capita income")
plt.title("Per capita income over the years")
plt.plot(df['year'], reg.predict(new_df), color='red',label='Predicted Per Capita Income')
plt.legend()
plt.show()
```
